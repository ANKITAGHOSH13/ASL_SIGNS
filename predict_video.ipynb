```python
import cv2
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from keras import utils
import time
video = cv2.VideoCapture(0)
total_frames = int(video.get(cv2.CAP_PROP_FRAME_COUNT))
frame_rate = video.get(cv2.CAP_PROP_FPS)
start_time = time.time()

while True:
    _, frame = video.read()

    # Convert BGR to RGB
    # image_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    # # plt.imshow(image_rgb)
    # print("image rgb")
    # print(image_rgb)

    frame_arr = utils.img_to_array(frame)
    
    # Resize the image
    resized_image = cv2.resize(frame_arr, (180, 180))
    # print("resized img")
    # print(resized_image)
    # plt.imshow(resized_image) #.numpy().astype(int))
    # plt.show()

    # Expand the dimensions from 3D to 4D (batch dimension)
    input_image = np.expand_dims(resized_image, 0)
    # print("input img")
    # print(input_image)

    # Make prediction
    prediction = np.array(model.predict(input_image))[0]
    print("pred")
    print(prediction)
    # Get predicted class index
    predicted_class_index = list(prediction).index(max(prediction))
    print("PCI")
    print(predicted_class_index)

    # Customize this part to your liking...
    if predicted_class_index == 0:
        print(f'Predicted class is Good')
    elif predicted_class_index == 1:
        print(f'Predicted class is Help')
    elif predicted_class_index == 2:
        print(f'Predicted class is Hungry')
    elif predicted_class_index == 3:
        print(f'Predicted class is Sorry')
    elif predicted_class_index == 4:
        print(f'Predicted class is Tired')

    cv2.imshow("Prediction", frame)

    key = cv2.waitKey(1)
    if key == ord('q'):
        break
end_time = time.time()

# Calculate total processing time
total_time = end_time - start_time

# Calculate frame rate
calculated_frame_rate = total_frames / total_time

print("Frame rate (from video metadata):", frame_rate)
print("Calculated frame rate:", calculated_frame_rate)

video.release()
cv2.destroyAllWindows()
```
